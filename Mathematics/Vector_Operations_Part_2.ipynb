### Importing the libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## Implementing the Dot Products

##### 2-D Points

```python
a = np.array([2,3])
b = np.array([4,4])

a_dot_b = np.dot(a,b)

print(a_dot_b)
```

##### 3-D Points

```python
a = np.array([2,3,4])
b = np.array([4,4,5])

a_dot_b = np.dot(a,b)

print(a_dot_b)
```

## Implementing the Cross Products

##### 2-D Points

```python
a = np.array([2,3])
b = np.array([4,4])

a_cross_b = np.cross(a,b)

print(a_cross_b)
```

##### 3-D Points

```python
a = np.array([2,3,4])
b = np.array([4,4,5])

a_cross_b = np.cross(a,b)

print(a_cross_b)
```

## Projection of Two Vectors

Projection Formula



Magnitude of Two Vectors:



```python
a = np.array([2,5])
v = np.array([8,-6])

mag_of_v = np.sqrt(sum(v**2))

print(mag_of_v)
```

```python
proj = (np.dot(a,v)/mag_of_v**2)*v
print(proj)
```
