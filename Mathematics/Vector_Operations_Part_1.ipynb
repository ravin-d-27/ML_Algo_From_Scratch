# Vector Operations in Python

## Basic Stuffs

### Importing the Libraries

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

### Providing a default theme to the plot

```python
sns.set()
```

### Plotting a Vector

```python
plt.quiver(0,0,4,5)
plt.show()
```

### Scaling the Vector and limits as well

```python
plt.quiver(0,0,4,5, scale_units = 'xy', angles= 'xy', scale = 1)
plt.xlim(-8,8)
plt.ylim(-8,8)
plt.show()
```

### Changing the Colour

```python
plt.quiver(0,0,4,5, scale_units = 'xy', angles= 'xy', scale = 1, color = 'r')
plt.quiver(0,0,-4,-5, scale_units = 'xy', angles= 'xy', scale = 1, color = 'b')
plt.plot(0,0,'ko') 
plt.xlim(-8,8)
plt.ylim(-8,8)
plt.show()
```

## Addition of Two Vectors

```python
v_1 = np.asarray([0,0,2,3])
v_2 = np.asarray([0,0,3,-2])

sum = v_1+v_2
print(sum)
```

```python
plt.quiver(0,0,2,3, scale_units = 'xy', angles= 'xy', scale = 1, color = 'r')
plt.quiver(0,0,3,-2, scale_units = 'xy', angles= 'xy', scale = 1, color = 'b')
plt.quiver(0,0,5,-1, scale_units = 'xy', angles= 'xy', scale = 1, color = 'g')
plt.plot(0,0,'ko') 
plt.xlim(-8,8)
plt.ylim(-8,8)
plt.show()
```

## Multiplying a Vector by a Scalar

### Positive Scalar Multiplication

```python
v_1 = np.asarray([0,0,2,3])
v_2 = 2*v_1

print(v_2)
```

```python
plt.quiver(0,0,2,3, scale_units = 'xy', angles= 'xy', scale = 1, color = 'r')
plt.quiver(0,0,4,6, scale_units = 'xy', angles= 'xy', scale = 1, color = 'g')
plt.plot(0,0,'ko') 
plt.xlim(-8,8)
plt.ylim(-8,8)
plt.show()
```

### Negative Scalar Multiplication

```python
v_1 = np.asarray([0,0,2,3])
v_2 = -0.5*v_1

print(v_2)
```

```python
plt.quiver(0,0,2,3, scale_units = 'xy', angles= 'xy', scale = 1, color = 'r')
plt.quiver(0,0,-1,-1.5, scale_units = 'xy', angles= 'xy', scale = 1, color = 'g')
plt.plot(0,0,'ko') 
plt.xlim(-8,8)
plt.ylim(-8,8)
plt.show()
```
